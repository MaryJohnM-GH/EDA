Aggregate Functions

```python
# Groupby
# Lets determine the mean price of each room_type
# Step 1 - group by room type
price_by_room_type = df.groupby('room_type')
price_by_room_type.first()
```

```python
# Step 2 get the mean price
price_by_room_type.price.mean()
```

```python
# Group by can be done using multiple columns also
# Lets group by room type and neighbourhood
multiple_group = df[:10].groupby(['room_type','neighbourhood'])
multiple_group.first()
```

Crosstab

```python
# Problem: see the room type distribution based on neighbourhood group
# generate a crosstab of room_type and neighbourhood_group
pd.crosstab(df['room_type'],df['neighbourhood_group'], margins = True)
```

PIVOT Table

```python
# Create a pivot table for neighbourhood_group and price
# Pivot table will return the mean of price by default
pd.pivot_table(df,index = ['neighbourhood_group'], values = "price")
```

```python
# Pivot table using multiple indexes
pd.pivot_table(df,index=['room_type','neighbourhood_group'],values='price')
```

```python
# Pivot table using multiple indexes
pd.pivot_table(df,index=['room_type','neighbourhood','neighbourhood_group'],values='price')
```

```python
# Pivot table using multiple indexes and using aggregate functions
pd.pivot_table(df,index=['room_type','neighbourhood','neighbourhood_group'],values='price',aggfunc=[np.mean,np.median,np.std,min,max])
```
