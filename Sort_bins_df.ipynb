```python
# Binning
# pandas.cut() and pandas.qcut are normally used
# pd.cut(x,bins,labels = None, right=true)
bins = [1979,1989,1999,2009]
labels = ['80s','90s','09s']
binned_years = pd.cut(df['Outlet_Establishment_Year'], bins,labels=labels, right=True)
print(binned_years)
```

```python
# . Using pandas.qcut()
# This method bins data into quantiles (bins with an equal number of data points).
# Syntax:
# pd.qcut(x, q, labels=None)
# •	x: The input array or column.
# •	q: The number of quantiles (bins) or a list of quantile edges.
# •	labels: Labels for the bins (optional).
df['Item_Weight_bin'] = pd.qcut(df['Item_Weight'], q=3, labels =['less weight','medium weight','heavy'])
df.head()
```
