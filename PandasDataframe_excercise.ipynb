Assignment 10: np.select
Hi there! I need a few columns created.

Create a ‘seasonal_bonus’ column that applies to these dates:
All days in December (month = 12)
Sundays (day_of_week = 6) in May (month = 5)
Mondays (day_of_week = 0) in July (month = 7)
Call the December bonus ‘Holiday Bonus’, the May bonus ‘Corporate Month’, and the July bonus ‘Summer Special’. If no bonus applies, the column should display ‘None’.
Finally, calculate the total bonus owed at $100 per day.

```python
# # Apply np.select
# result = np.select(conditions, choices, default=30)

conditions = [
    transactions_df['month'] == 12,
    (transactions_df['month'] == 5) & (transactions_df['day'] == 6),
    (transactions_df['month'] == 7) & (transactions_df['day'] == 0)
]
choices = ['Holiday Bonus', 'Corporate Month','Summer Special']

transactions_df['seasonal_bonus'] = np.select(conditions,choices, default = 'None')
transactions_df.sample(10)
```

```python
transactions_df['seasonal_bonus'].isin(['Holiday Bonus', 'Corporate Month','Summer Special']).count() * 100
```
